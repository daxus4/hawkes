```python
import pandas as pd
import numpy as np
import os
```

```python
files = []
for filename in os.listdir('/home/davide/Desktop/phd/bitfinex-api-py/data/orderbook_changes'):
    if filename.endswith('.tsv'):
        files.append('/home/davide/Desktop/phd/bitfinex-api-py/data/orderbook_changes' + filename)
files[0]
```

```python
dfs = []
for f in files:
    print(f)
    df = pd.read_csv('/home/davide/Desktop/phd/bitfinex-api-py/data/orderbook_changes/orderbook_changes_1705168243495.tsv', sep='\t')
    df['MidPrice'] = (df["AskPrice1"]+df["BidPrice1"])/2
    df['Return'] = (-df["MidPrice"]+df["MidPrice"].shift(-1)) / df["MidPrice"]

    for base_imbalance_orderbook_level in range(2, 15):
        pbid = df["BidPrice1"] - df[f"BidPrice{base_imbalance_orderbook_level}"]
        pask = df[f"AskPrice{base_imbalance_orderbook_level}"] - df["AskPrice1"]
        df[f"BaseImbalance{base_imbalance_orderbook_level}"] = (pbid-pask)/(pbid+pask)

    df=df.dropna(subset=(['Return'] + [f'BaseImbalance{i}' for i in range(2, 15)]))
    df = df[df['Return'] != 0]
    df = df[ [f'BaseImbalance{i}' for i in range(2, 15)] + ['Return']]
    dfs.append(df)

df = pd.concat(dfs)
df
```

```python
corrs = []
for base_imbalance_orderbook_level in range(2, 15):
    corrs.append(df['Return'].corr(df[f'BaseImbalance{base_imbalance_orderbook_level}'], method='spearman'))

corrs
```

```python
corrs = []
for base_imbalance_orderbook_level in range(2, 15):
    corrs.append(df['Return'].corr(df[f'BaseImbalance{base_imbalance_orderbook_level}'], method='pearson'))

corrs
```
